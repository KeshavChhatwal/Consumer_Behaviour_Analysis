### Mapping the items purchased to the relevant category

```python
correct_mapping= {
    
# Clothing
"T-shirt": "Clothing", "Shirt": "Clothing", "Shorts": "Clothing", "Hoodie": "Clothing",
"Pants": "Clothing", "Socks": "Clothing",
"Jeans": "Clothing", "Blouse": "Clothing", "Skirt": "Clothing",
"Sweater": "Clothing", "Dress": "Clothing",

# Accessories
"Sunglasses": "Accessories", "Gloves": "Accessories", "Jewelry": "Accessories", "Hat": "Accessories" , "Handbag": "Accessories",
"Backpack": "Accessories" , "Belt": "Accessories", "Scarf": "Accessories",
"Bag": "Accessories",

# Electronics
"Laptop": "Electronics", "Phone": "Electronics",
"Headphones": "Electronics", "Watch": "Electronics",

# Footwear
"Shoes": "Footwear", "Sandals": "Footwear",
"Sneakers": "Footwear", "Boots": "Footwear",

# Outerwear
"Coat": "Outerwear", "Jacket": "outerwear" 
}
```

```python
correct_mapping
```

### Replacing the values and fixing it in the original data after mapping.

```python
df['Category']=df['Item Purchased'].map(correct_mapping)
```

```python
df.groupby('Category')['Item Purchased'].unique()
```

# Treating Null Values

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.groupby('Category')['Size'].unique()
# nan represents the null value in each category therefore we will treat them accordingly
```

```python
df.loc[df['Category']=='Electronics','Size']='Not Applicable'
```

```python
df.loc[df['Category']=='Accessories','Size']='Free Size'
```

```python
df.loc[df['Category']=='Footwear','Size']='Not Available'
```

```python
# Accessories and Electronics are the products which does not fall in the class of sizing so we will replace the values with not applicable.
# Footwear sizes are not mesured in "M, L, S, XL" hence will treat them as well.
df.groupby('Category')['Size'].unique()
```

```python
# Treating Null Values for Clothng 
df.loc[(df['Category'] == 'Clothing') & (df['Size'].isnull()), 'Size'] = df[df['Category'] == 'Clothing']['Size'].mode()[0]
```

```python
df.isnull().sum()
```

### Treating Null Values for Review Rating

```python
df['Review Rating'].mean()
```

```python
df['Review Rating'].median()
```

```python
df['Review Rating'].mode()
```

```python
df.groupby("Item Purchased")["Review Rating"].mean()
```

```python
# Filling the null values of review rating with the mean of each item purchased 
df['Review Rating']= df['Review Rating'].fillna(df.groupby("Item Purchased")['Review Rating'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
# df.head(10)
```

```python
df['Review Rating']=df['Review Rating'].round(2)
```

```python
df.head(10)
```

### Treating Null Values of Previous Purchases

```python
df.isnull().sum()
```

```python
df["Previous Purchases"]=df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df[df['Category'].isnull()]['Item Purchased'].unique()
```

### Treating Null Values of Purchase Amount (USD)

```python
df.groupby("Item Purchased")["Purchase Amount (USD)"].mean()
```

```python
df['Purchase Amount (USD)']= df['Purchase Amount (USD)'].fillna(df.groupby("Item Purchased")['Purchase Amount (USD)'].transform('mean'))
```

```python
df.isnull().sum()
```
